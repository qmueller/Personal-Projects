```python
import pandas as pd
from wordcloud import WordCloud
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
costume_df = pd.read_csv(r'C:\Users\quinn\OneDrive\Documents\Quinn Python Scripts\upload to python\Halloween_Costumes_2023.csv')
```

```python
costume_df
```

```python
dictfreq = dict(zip(costume_df.Costume,costume_df.Frequency))
```

```python
pumpkin_png = np.array(Image.open(r'C:\Users\quinn\OneDrive\Documents\Quinn Python Scripts\upload to python\pumpkin.jpg')) 
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='black',
                prefer_horizontal=0.9,  
                colormap='Set2', 
                mask=pumpkin_png,
                contour_width=10,
                contour_color='orange',      
                min_font_size = 10,
                random_state = 2023)
wordcloud.generate_from_frequencies(frequencies=dictfreq) 
plt.figure(figsize = (5, 5)) 
plt.imshow(wordcloud) 
plt.axis("off")
plt.savefig('Costumes.jpg', dpi=300)
plt.show() 
```
